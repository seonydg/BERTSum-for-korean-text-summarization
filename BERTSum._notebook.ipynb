## 원본 데이터에서 필요 데이터 추출
- 'text'와 'extractive' 추출
- 신문기사, 기고문, 법원 판결문 3개로 나누어져 있는 파일을 하나로 합침

### train dataset

```python
data_path = '/content/drive/MyDrive/NLP Dataset/문서요약 텍스트(AI Hub)/raw_data/train'
filenames = [x for x in os.listdir(data_path) if x.endswith('json')]
filenames.sort()
filenames
```

```python
list_dic = []

for file in filenames:
  filelocation = os.path.join(data_path, file)

  with open(filelocation, 'r') as json_file:
    data = json.load(json_file)['documents']

    for x in tqdm (range(len(data))):
      text = data[x]['text']
      text = str(text).replace('"', "'")

      extractive = data[x]['extractive']
      for index, value in enumerate(extractive):
        if value == None:
          extractive[index] = 0

      p = re.compile('(?<=sentence\'\: \')(.*?)(?=\'highlight_indices)')
      texts = p.findall(text)

      sentences = []
      for t in texts:
        sentence = t[:-3]
        sentences.append(sentence)

      mydict = {}
      mydict['text'] = sentences
      mydict['extractive'] = extractive
      list_dic.append(mydict)
```

- 코랩에서 RAM 문제로 train dataset 분할할 필요가 있음

```python
def list_chunk(lst, n):
    return [lst[i:i+n] for i in range(0, len(lst), n)]

data_chunked = list_chunk(list_dic, len(list_dic)//10) # 10개로 분할

for i, d in enumerate(data_chunked):
    with open('/content/drive/MyDrive/NLP Dataset/문서요약 텍스트(AI Hub)/raw_data/train.{}.json'.format(i), 'w') as fh:
        json.dump(d, fh)
```

## 학습 가능한 형태로 데이터 변환

## Train Dataset

- Source 및 Target 설정
  - Source; 원본 데이터의 'text'에 해당
  - Target; 원본 데이터의 'extractive' 리스트의 index에 해당하는 문장을 source에서 추출
- Source 및 Target 을 Tokenizing
  - Tokenizer; MeCab + BertTokenizer
    - 한국어 데이터이므로 MeCab으로 형태소 분석 후 BertTokenizer 사용
  - Tokenizing 후 token 단위로 source 및 target 저장

```python
data_path = '/content/drive/MyDrive/NLP Dataset/문서요약 텍스트(AI Hub)/raw_data'
filenames = [x for x in os.listdir(data_path) if 'train' in x and x.endswith('json')]
filenames
```
